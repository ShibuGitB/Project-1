***So droping the remaining columns except critic_score, na_sales (sales in north america), jp_sales (sales in japan), pal_sales (sales in europe,africa), other_sales, total_sales***

```python
X.drop(["img","title","console","genre","publisher","developer","release_date","last_update"],axis=1,inplace=True)
X
```

***Re-arranged input & output in same order***

```python
input1=X.iloc[:,0:1]
input2=X.iloc[:,2:]
input3=X.iloc[:,1:2]
ds=pd.concat([input1,input2],axis=1)
X=pd.concat([ds,input3],axis=1)
X
```

***Checking if there's a Missing Values or not***

```python
print (X.isna().sum())
```

***Handling the missing values by mean & median***

```python
a=X["critic_score"].mean()
X["critic_score"]=X["critic_score"].fillna(a)
```

```python
b=X["total_sales"].mean()
X["total_sales"]=X["total_sales"].fillna(b)
```

```python
c=X["na_sales"].mean()
X["na_sales"]=X["na_sales"].fillna(c)
```

```python
d=X["jp_sales"].mean()
X["jp_sales"]=X["jp_sales"].fillna(d)
```

```python
e=X["pal_sales"].mean()
X["pal_sales"]=X["pal_sales"].fillna(e)
```

```python
f=X["other_sales"].mean()
X["other_sales"]=X["other_sales"].fillna(f)
```

***Selected & Seperated input & output from dataset for Total Sales prediction(Machine Learning) process***

***Seperated & stored input in x variable***

```python
x=X.iloc[:,0:5]
x
```

***Seperated & stored output in y variable***

```python
y=X.iloc[:,5]
y
```

***Scaling input by using Min Max Scaler***

```python
so=MinMaxScaler()
so.fit(x)
si=so.transform(x)
```

***Tooks seperated input & output for Training & Testing***

```python
x_train,x_test,y_train,y_test=train_test_split(si,y,test_size=0.3,random_state=1)
```

***Applied Multiple-Linear-Regrssion algorithm for the Machine Learning Process and made a Model***

*Takes multiple linear regression over simple linear regression beacuse of our data set have multiple inputs,whwn it had single input & output process based on the simple linear regression vice-versa here we have multiple input so we selected MULTIPLE-LINEAR-REGRESSION*

```python
z=LinearRegression()
z.fit(x_train,y_train)
```

***With the Y-Pred(predicted output by X-Test) and Y-Test(actual output of the X-Test) predicting the regression accuracy score R2 SCORE***

```python
w=z.predict(x_test)
q=y_test.values
```

```python
print (r2_score(q,w))
```

***Applied another Regression algorithm - KNearestNeighbors for Regression(KNN) and made a model for the ML process***

```python
from sklearn.neighbors import KNeighborsRegressor
obj=KNeighborsRegressor(n_neighbors=3)
obj.fit(x_train,y_train)
```

***With the Y-Pred(predicted output by X-Test) and Y-Test(actual output of the X-Test) predicting the regression accuracy score R2 SCORE***

```python
o=obj.predict(x_test)
i=y_test.values
```
